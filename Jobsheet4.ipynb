Praktikum 1: Inheritance Dasar

```python
class Kendaraan:
    def __init__(self, merk):
        self.merk = merk

    def mulai_mesin(self):
        print(f"Mesin kendaraan {self.merk} dinyalakan.")

class Mobil(Kendaraan):
    def __init__(self, merk, warna):
        self.merk = merk
        self.warna = warna

    def info_mobil(self):
        print(f"Ini adalah mobil {self.merk} berwarna {self.warna}.")

# Main
mobil_tesla = Mobil("Tesla Model S", "Merah")
mobil_tesla.info_mobil()
mobil_tesla.mulai_mesin()
print(mobil_tesla.merk)
```

Praktikum 2: super()

```python
class Person:
    def __init__(self, nama, usia):
        self.nama = nama
        self.usia = usia

    def perkenalkan_diri(self):
        print(f"Halo, nama saya {self.nama}, usia {self.usia}")

class Student(Person):
    def __init__(self, nama, usia, student_id, jurusan):
        super().__init__(nama, usia)
        self.student_id = student_id
        self.jurusan = jurusan

    def perkenalkan_diri(self):
        super().perkenalkan_diri()
        print(f"Saya mahasiswa {self.jurusan} dengan ID {self.student_id}")

mahasiswa = Student("Dewi", 20, "M001", "Teknik Komputer")
mahasiswa.perkenalkan_diri()
```

Praktikum 3: Method Overriding

```python
class Hewan:
    def __init__(self, nama):
        self.nama = nama

    def bersuara(self):
        print("Suara hewan")

class Kucing(Hewan):
    def __init__(self, nama, ras):
        super().__init__(nama)
        self.ras = ras

    def bersuara(self):
        print(f"{self.nama} mengeong")

kucing = Kucing("Puspus", "Persia")
kucing.bersuara()
```

Praktikum 4: Multi-level Inheritance

```python
class Organisme:
    def __init__(self, nama):
        self.nama = nama

class Hewan(Organisme):
    def __init__(self, nama, makanan):
        super().__init__(nama)
        self.makanan = makanan

class Mamalia(Hewan):
    def __init__(self, nama, makanan, kaki):
        super().__init__(nama, makanan)
        self.kaki = kaki

kucing = Mamalia("Kucing", "Daging", 4)
print(kucing.nama)
```

Praktikum 5: Hierarchical Inheritance

```python
class Pegawai:
    def __init__(self, nama, gaji):
        self.nama = nama
        self.gaji = gaji

    def hitung_gaji(self):
        return self.gaji

class Manager(Pegawai):
    def __init__(self, nama, gaji, tunjangan):
        super().__init__(nama, gaji)
        self.tunjangan = tunjangan

    def hitung_gaji(self):
        return self.gaji + self.tunjangan

class Staf(Pegawai):
    def __init__(self, nama, gaji, bonus):
        super().__init__(nama, gaji)
        self.bonus = bonus

    def hitung_gaji(self):
        return self.gaji + self.bonus


manager = Manager("Budi", 10000000, 5000000)
staf = Staf("Citra", 7000000, 1500000)

print("Gaji Manager:", manager.hitung_gaji())
print("Gaji Staf:", staf.hitung_gaji())
```

Program Hierarki Mahasiswa

```python
class Mahasiswa:
    def __init__(self, nama, nim, ipk):
        self.nama = nama
        self.nim = nim
        self.ipk = ipk

    def tampilkan_info(self):
        print(f"Nama: {self.nama}")
        print(f"NIM: {self.nim}")
        print(f"IPK: {self.ipk}")

    def hitung_predikat(self):
        if self.ipk >= 3.5:
            print("Dengan Pujian")
        elif self.ipk >= 3.0:
            print("Sangat Memuaskan")
        elif self.ipk >= 2.5:
            print("Memuaskan")
        else:
            print("Cukup")

class MahasiswaSarjana(Mahasiswa):
    def __init__(self, nama, nim, ipk, semester, sks):
        super().__init__(nama, nim, ipk)
        self.semester = semester
        self.sks = sks

    def tampilkan_info(self):
        super().tampilkan_info()
        print(f"Semester: {self.semester}")
        print(f"SKS Lulus: {self.sks}")

class MahasiswaMagister(Mahasiswa):
    def __init__(self, nama, nim, ipk, tesis, pembimbing):
        super().__init__(nama, nim, ipk)
        self.tesis = tesis
        self.pembimbing = pembimbing

    def tampilkan_info(self):
        super().tampilkan_info()
        print(f"Judul Tesis: {self.tesis}")
        print(f"Pembimbing: {self.pembimbing}")

# Main
mhs1 = MahasiswaSarjana("Andi", "123", 3.6, 6, 120)
mhs2 = MahasiswaMagister("Budi", "456", 3.4, "AI", "Dr. Siti")

mhs1.tampilkan_info()
mhs1.hitung_predikat()

print("-----")

mhs2.tampilkan_info()
mhs2.hitung_predikat()
```
