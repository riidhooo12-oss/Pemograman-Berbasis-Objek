PRAKTIKUM 01 – Polimorfisme (Inheritance & Overriding)

```
class Burung:
    def __init__(self, nama):
        self.nama = nama

    def terbang(self):
        print(f"{self.nama} terbang dengan cara umum.")

    def bersuara(self):
        print(f"{self.nama} mengeluarkan suara burung.")


class Elang(Burung):
    def __init__(self, nama, rentang_sayap):
        super().__init__(nama)
        self.rentang_sayap = rentang_sayap

    def terbang(self):
        print(f"{self.nama} terbang tinggi melayang di angkasa.")

    def bersuara(self):
        print(f"{self.nama} berteriak nyaring!")


class Pipit(Burung):
    def __init__(self, nama, warna_bulu):
        super().__init__(nama)
        self.warna_bulu = warna_bulu

    def terbang(self):
        print(f"{self.nama} terbang cepat di antara pepohonan.")

    def bersuara(self):
        print(f"{self.nama} berkicau merdu: Cit cit!")


def demonstrasi_aksi_burung(daftar):
    for burung in daftar:
        burung.terbang()
        burung.bersuara()


if __name__ == "__main__":
    data = [
        Elang("Elang Jawa", 1.5),
        Pipit("Pipit Gereja", "Coklat"),
        Burung("Burung Biasa")
    ]
    demonstrasi_aksi_burung(data)
```

PRAKTIKUM 02 – Duck Typing

```
class Kucing:
    def bersuara(self):
        print("Kucing: Meow!")

class Anjing:
    def bersuara(self):
        print("Anjing: Guk guk!")

class Bebek:
    def bersuara(self):
        print("Bebek: Kwek kwek!")

class Mobil:
    def jalankan(self):
        print("Mobil: Brummm!")

def buat_suara(obj):
    try:
        obj.bersuara()
    except:
        print("Objek tidak bisa bersuara")


if __name__ == "__main__":
    data = [Kucing(), Anjing(), Bebek(), Mobil()]
    for item in data:
        buat_suara(item)
```

PRAKTIKUM 03 – Overloading (args)

```
class KalkulatorSederhana:
    def tambah(self, *args):
        total = 0
        for angka in args:
            if isinstance(angka, (int, float)):
                total += angka
        print("Hasil:", total)


if __name__ == "__main__":
    k = KalkulatorSederhana()
    k.tambah(1, 2)
    k.tambah(1, 2, 3, 4)
    k.tambah()
```

PRAKTIKUM 04 – Overloading (*args & kwargs)

```
class DataProcessor:
    def process(self, *args, **kwargs):
        print("Args:", args)
        print("Kwargs:", kwargs)


if __name__ == "__main__":
    dp = DataProcessor()
    dp.process(1, 2, 3)
    dp.process(a=10, b=20)
    dp.process(1, 2, x=100, y=200)
```

PRAKTIKUM 05 – Kombinasi Polimorfisme

```
import math

class Bentuk:
    def hitung_luas(self):
        raise NotImplementedError


class Lingkaran(Bentuk):
    def __init__(self, r):
        self.r = r

    def hitung_luas(self):
        return math.pi * self.r ** 2


class Persegi(Bentuk):
    def __init__(self, s):
        self.s = s

    def hitung_luas(self):
        return self.s * self.s


class TaplakMeja:
    def __init__(self, p, l):
        self.p = p
        self.l = l

    def hitung_luas(self):
        return self.p * self.l


def tampilkan(obj):
    try:
        print("Luas:", obj.hitung_luas())
    except:
        print("Tidak bisa dihitung")


if __name__ == "__main__":
    data = [
        Lingkaran(7),
        Persegi(5),
        TaplakMeja(2, 3),
        "string"
    ]
    for d in data:
        tampilkan(d)
```

PRAKTIKUM 06 – isinstance()

```
class Burung:
    def __init__(self, nama):
        self.nama = nama

    def terbang(self):
        print(self.nama, "terbang")

    def bersuara(self):
        print(self.nama, "bersuara")


class Elang(Burung):
    def berburu(self):
        print(self.nama, "berburu")


class Pipit(Burung):
    def membangun_sarang(self):
        print(self.nama, "buat sarang")


def interaksi(obj):
    if isinstance(obj, Burung):
        obj.terbang()
        obj.bersuara()

        if isinstance(obj, Elang):
            obj.berburu()
        elif isinstance(obj, Pipit):
            obj.membangun_sarang()
    else:
        print("Bukan burung")


if __name__ == "__main__":
    data = [Elang("Elang"), Pipit("Pipit"), Burung("Merpati"), "kucing"]
    for d in data:
        interaksi(d)
```

Penugasan

```
class Komputer:
    def __init__(self, merk, processor, ram_gb):
        self.merk = merk
        self.processor = processor
        self.ram_gb = ram_gb

    def info_spesifikasi(self):
        print(f"Merk: {self.merk}")
        print(f"Processor: {self.processor}")
        print(f"RAM: {self.ram_gb} GB")

    def jalankan_aplikasi(self, app):
        print(f"{self.merk} menjalankan {app}")

    # Overloading (default argument)
    def upgrade_ram(self, tambah, tipe="DDR4"):
        self.ram_gb += tambah
        print(f"Upgrade {tambah} GB ({tipe}) -> Total: {self.ram_gb} GB")


class Laptop(Komputer):
    def __init__(self, merk, processor, ram_gb, layar, berat):
        super().__init__(merk, processor, ram_gb)
        self.layar = layar
        self.berat = berat

    def info_spesifikasi(self):
        super().info_spesifikasi()
        print(f"Layar: {self.layar} inch")
        print(f"Berat: {self.berat} kg")


class Desktop(Komputer):
    def __init__(self, merk, processor, ram_gb, casing, monitor):
        super().__init__(merk, processor, ram_gb)
        self.casing = casing
        self.monitor = monitor

    def info_spesifikasi(self):
        super().info_spesifikasi()
        print(f"Casing: {self.casing}")
        print(f"Monitor: {'Ya' if self.monitor else 'Tidak'}")


def cetak_semua(data):
    for d in data:
        print("------")
        d.info_spesifikasi()


if __name__ == "__main__":
    l = Laptop("Asus", "i5", 8, 14, 1.5)
    d = Desktop("HP", "i7", 16, "Tower", True)

    semua = [l, d]

    cetak_semua(semua)

    l.jalankan_aplikasi("Chrome")

    # Overloading
    l.upgrade_ram(4)
    d.upgrade_ram(8, "DDR5")
```
