Praktikum 04 Method

```python
class Mobil:
    def __init__(self, merk, harga):
        self.merk = merk
        self.harga = harga

    def diskon(self, persen):
        return self.harga - (self.harga * persen / 100)

m = Mobil("Toyota", 300000000)
print(m.diskon(10))
```

Penugasan 01

```python
%%writefile parkir.py
# Modul parkir

def hitung_biaya(jam):
    tarif_per_jam = 2000
    return jam * tarif_per_jam
```

```python
import parkir

class Kendaraan:
    def __init__(self, plat, jenis):
        self.plat = plat
        self.jenis = jenis
        self.jam_masuk = 0
        self.jam_keluar = 0

    def masuk(self, jam):
        self.jam_masuk = jam
        print(f"Kendaraan {self.plat} masuk jam {jam}")

    def keluar(self, jam):
        self.jam_keluar = jam
        lama = self.jam_keluar - self.jam_masuk
        biaya = parkir.hitung_biaya(lama)

        print(f"Kendaraan {self.plat} keluar jam {jam}")
        print(f"Lama parkir: {lama} jam")
        print(f"Total biaya: Rp {biaya}")


# Program utama4
def main():
    plat = input("Masukkan plat nomor: ")
    kendaraan = Kendaraan(plat, "Motor")

    jam_masuk = int(input("Jam masuk: "))
    kendaraan.masuk(jam_masuk)

    jam_keluar = int(input("Jam keluar: "))
    kendaraan.keluar(jam_keluar)


if __name__ == "__main__":
    main()
```
